### La simulaci√≥n de montecarlo

* Generamos dos n√∫mero aleatorios x e y entre 0 y 1
* Calcularemos x^2+y^2
    * Si el valor es inferior a 1 estamos dentro del c√≠rculo
    * Si el valor es superior a 1 estamos fuera del c√≠rculo
* Calculamos el n√∫mero total de veces que cae dentro del c√≠rculo y lo dividimos entre el n√∫mero total de intentos para obtener una aproximaci√≥n de la probabilidad de caer dentro del c√≠rculo
* Usamos dicha probabilidad para aproximar el valor de \pi
* Repetimos el experimento un n√∫mero de veces (ejemplo 1000), para obtener (1000) diferentes aproximaciones de \pi
* Calculamos el promedio de los experimentos anteriores para un valor final de \pi

```python
def pi_montecarlo(n,n_exp):
    '''n: N√∫mero de intentos en cada experimento
    n_exp: N√∫mero de experimentos'''
    pi_avg=0 
    pi_value_list=[]
    for i in range(n_exp): #N√∫mero de veces que se repite el experimento 
        value=0
        x=np.random.uniform(0,1,n).tolist() #Todos los puntos tienen igual probabilidad
        y=np.random.uniform(0,1,n).tolist() #Todos los puntos tienen igual probabilidad
        for j in range(n):
            z=np.sqrt((x[j]**2)+(y[j]**2))
            if z<=1:
                value+=1
        float_value=float(value)
        pi_value=(float_value*4)/n
        pi_value_list.append(pi_value)
        pi_avg+=pi_value
    pi=pi_avg/n_exp
    print("El valor de pi es: "+str(pi))
    fig=plt.plot(pi_value_list)
    return (pi,fig)
```

```python
pi_montecarlo(1000,400)
```
