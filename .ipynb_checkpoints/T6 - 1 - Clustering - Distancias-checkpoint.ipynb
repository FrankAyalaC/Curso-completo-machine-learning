# Clustering JerÃ¡rquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage,dendrogram
```

```python
data
```

```python
Z_ward=linkage(data,"ward")
Z_mean=linkage(data,"average")
```

```python
fig,[ax1,ax2]=plt.subplots(1,2)
fig.suptitle("Dendograma jerÃ¡rquico para el clustering")

ax1.set_title("MÃ©todo de enlace Ward")
ax1.set_xlabel('ID de los usuarios')
ax1.set_ylabel("Distancia")
# dendrogram(Z,leaf_rotation=90,leaf_font_size=7)

ax2.set_title("MÃ©todo de enlace Promedio")
ax2.set_xlabel('ID de los usuarios')
ax2.set_ylabel("Distancia")

dendrogram(Z_ward,leaf_rotation=90,leaf_font_size=7,ax=ax1)
dendrogram(Z_mean,leaf_rotation=90,leaf_font_size=7,ax=ax2)
```

```python
Z_min=linkage(data,"single")
Z_max=linkage(data,"complete")
```

```python
fig,[ax1,ax2]=plt.subplots(1,2)
fig.suptitle("Dendograma jerÃ¡rquico para el clustering")

ax1.set_title("MÃ©todo de enlace Single")
ax1.set_xlabel('ID de los usuarios')
ax1.set_ylabel("Distancia")
# dendrogram(Z,leaf_rotation=90,leaf_font_size=7)

ax2.set_title("MÃ©todo de enlace Complete")
ax2.set_xlabel('ID de los usuarios')
ax2.set_ylabel("Distancia")

dendrogram(Z_min,leaf_rotation=90,leaf_font_size=7,ax=ax1)
dendrogram(Z_max,leaf_rotation=90,leaf_font_size=7,ax=ax2)
```
