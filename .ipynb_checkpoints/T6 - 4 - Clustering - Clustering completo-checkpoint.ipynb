# Clustering con Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../GitHub/python-ml-course/datasets/wine/winequality-red.csv',sep=';')
df.head()
```

```python
df.shape
```

```python
plt.hist(df.quality)
```

```python
df.groupby("quality").mean()
```

## Normalización de los datos

```python
df_norm=(df-df.min())/(df.max()-df.min())
df_norm
```

# Clustering jerárquico con scikit-learn

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
clus=AgglomerativeClustering(n_clusters=6,linkage="ward").fit(df_norm)
```

```python
md_h=pd.Series(clus.labels_)
plt.hist(md_h)
plt.title("Histograma de los clusters")
plt.xlabel('ID del cluster')
plt.ylabel('Número de vinos en cada cluster')
```

Estos son todos y cada uno de los hijos

```python
clus.children_
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
Z=linkage(df_norm,"ward")
```

```python
plt.figure(figsize=(11,8))
plt.title("Dendograma para el clustering jerárquico truncado")
plt.xlabel('ID del vino')
plt.ylabel("Distancia")
dendrogram(Z,leaf_rotation=90,leaf_font_size=12,truncate_mode="lastp",
           p=6,show_leaf_counts=True,show_contracted=True)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
model=KMeans(n_clusters=6)
model.fit(df_norm)
```

```python
model.labels_
```

```python
md_k=pd.Series(model.labels_)
```

```python
df_norm["clust_h"]=md_h
df_norm["clust_k"]=md_k
df_norm.head()
```

```python
md_h=pd.Series(clus.labels_)

plt.figure(figsize=(12,6))

plt.suptitle('Histograma de los clusters')

plt.subplot(121)
plt.hist(md_h)
plt.title("Método jerárquico")
plt.xlabel('ID del cluster')
plt.ylabel('Número de vinos en cada cluster')

plt.subplot(122)
plt.hist(md_k,color='g')
plt.title("Método K-means")
plt.xlabel('ID del cluster')
plt.ylabel('Número de vinos en cada cluster')

plt.subplots_adjust(wspace=0.4)
```

```python
baricentro_K=pd.DataFrame(model.cluster_centers_,columns=df_norm.columns.values.tolist()[:-2])
```
