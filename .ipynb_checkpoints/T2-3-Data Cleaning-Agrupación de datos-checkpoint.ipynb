# Agrupación de datos por categoría

Variable categorica se usa para juntar información de una misma categoría y concluir a partir de dicha categoría{

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gender=["Female","Male"]
income=["Poor","Middle Class","Rich"]
```

```python
gender_data=[]
income_data=[]
n=500
for i in range (n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
income_data[:10]
```

```python
height=170+30*np.random.randn(n)
weight=65+25*np.random.rand(n)
age=30+12*np.random.rand(n)
income=18000+3500*np.random.rand(n)
```

```python
data=pd.DataFrame({"Gender":gender_data,"Economic Status":income_data,"Height":height,
                  "Weight":weight,"Age":age,"Income":income})
```

```python
data
```

Al DataFrame se le aplica df.Groupby con el nombre de la(s) columna(s) de interés.

Esta función crea un objeto groubpy que tendrá 2 atributos: el nombre y el grupo. Se puede acceder a ellos a través de la sintaxis de puntos. El nombre son las categorías existentes y el grupo es el conjunto que obedece a cada nombre o categoría

```python
grouped_gender=data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names,groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group=data.groupby(["Gender","Economic Status"])
len(double_group) #Devuelve el número de grupos
```

```python
double_group.head()
```

```python
for names,groups in double_group:
    print(names)
    print(groups)
```

## Agregado de datos

Este método permite agregar cosas específicas en nuestro data set

```python
double_group.aggregate({"Income":np.sum,"Age":np.mean,"Height":np.std})
```

```python
double_group.aggregate({"Age":np.mean,"Height":lambda h:np.mean(h)/np.std(h)}) #Tipificación para la altura
```

```python
double_group.aggregate((np.sum,np.mean,np.std))
```

```python
double_group.aggregate((lambda x:np.mean(x)/np.std(x)))
```

## Filtrado de datos

La funcion filter se usa después de elegir la variable a filtrar y devuelve los elementos del grupo que cumplen determinado criterio.

```python
double_group["Age"].filter(lambda x:x.sum()>3100)
```

```python
double_group.sum()
```

## Transformación de variables

A partir de los grupos formados por las variables categóricas, se determina la funcion **.transform()**, dado que los criterios empleados son sobre variables numéricas.

```python
zscore=lambda x:(x-x.mean())/x.std() # Cada x representa una columna
```

```python
z_group=double_group.transform(zscore)
```

```python
z_group
```

```python
plt.hist(z_group["Age"])
```

```python
fill_na_mean=lambda x:x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```

Un data set original se divide por lo regular en 2 partes:

* Conjunto de training o entrenamiento: Con dicha información se crea el modelo predictivo (75-80%)
* Conjunto de testing: Para evaluar la eficacia del modelo (25-20%)

Es posible que un modelo sirva muy bien para el training apero muy mal en el conjunto de testing.

## Conjunto de entrenamiento y de testing

```python
data=pd.read_csv('../GitHub/python-ml-course/datasets/customer-churn-model/Customer Churn Model.txt')
```

### Dividir el conjunto utilizando una distribución normal

```python
a=np.random.randn(len(data))
```

```python
plt.hist(a)
```

```python
# calculando el percentil 80 del vector a
perc_80 = np.percentile(a, 80)

# creando el vector check
check = (a<perc_80) #Porcentaje de los datos
```

```python
plt.hist(check*1) #Un vector booleano se multiplica por 1 para convertirlo a ceros y unos
```

```python
training=data[check]
testing=data[~check] #El símbolo ~cambia los valores del vector booleano
```

```python
~check
```

```python
a=len(training)
b=len(testing)
print(a,b,a+b)
```

### Con la librería Sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(data,test_size=0.2)
```

```python
a=len(train)
b=len(test)
print(a,b,a+b)
```
