# Enfoques usados para el manejo de datos faltantes

## Enfoque más drástico: Borrado de datos

Se eliminan las filas o columnas donde haya datos faltantes.

```python
df_titanic.dropna(axis=0,how="all") 
#0 eje de filas y 1 igual a columnas, how="any" or "all" si todas o alguna columna tiena na
```

```python
df_titanic2=df_titanic
df_titanic2.dropna(axis=0,how="any") 
```

## Enfoque menos drástico: Cómputo de los valores faltantes

Existen diferentes alternativas: llenar con un número específico, con un promedio, o demás. El llenado dependerá del contexto en que se esté trabajando.

```python
df_titanic3=df_titanic
a=0 #Puede ser un string, int, float
df_titanic3=df_titanic3.fillna(a) 
df_titanic3
```

Se puede llenar dependiendo de la columna, así se realiza una correspondencia entre los formatos de la columna original y los reemplazos.

```python
df_titanic4=df_titanic
df_titanic4.body=df_titanic4.body.fillna(0) #Se debe sobreescribir la columna
df_titanic4["home.dest"]=df_titanic4["home.dest"].fillna("desconocido")
```

```python
df_titanic4
```

### Llenando con el promedio

```python
df_titanic5=df_titanic
df_titanic5.age.fillna(df_titanic5.age.mean()) #Puede usarse la media, mediana, funciones lambda, etc.
```

El método fillna tiene un forward fill (ffill) y un backward fill (backfill) que como su nombre lo indica permite que llenemos el dato faltante con el primero que esté adelante o atrás.

```python
df_titanic5.age[1290],df_titanic5.age[1291],df_titanic5.age[1292]
```

```python
df_titanic5.age.fillna(method="ffill")[1291]
```

# Variables dummy (variables de ayuda, artificiales)

### Dumificación

Todas las variables serán cero menos una que lleva la información de la categoría.

```python
dummy_sex=pd.get_dummies(df_titanic.sex,prefix="sex")
```

```python
dummy_sex.head(10)
```

Para adicionarle estas variables dummies se puede:

- Eliminar la variable categorica

```python
column_name=df_titanic.columns.values.tolist()
```

```python
column_name
```

```python
df_titanic6=df_titanic.drop(["sex"],axis=1)
df_titanic6
```

```python
df_titanic6=pd.concat([df_titanic6,dummy_sex],axis=1)
df_titanic6
```

Se puede hacer todo de golpe

```python
def CreateDummies(df,var_name):
    dummy=pd.get_dummies(df[var_name],prefix=var_name)
    df=df.drop(var_name,axis=1)
    df=pd.concat([df,dummy],axis=1)
    
    return df
```
